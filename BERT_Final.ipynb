```python
import os, sys, shutil
import time
import gc
from contextlib import contextmanager
from pathlib import Path
import random
import numpy as np, pandas as pd
from tqdm import tqdm, tqdm_notebook
```

```python
train_raw = pd.read_csv('train_reduced.csv')
test_raw = pd.read_csv('test.csv')
```

```python
train_raw
```

```python
train = pd.DataFrame(columns=['text_a', 'text_b', 'labels'])
train['text_a'] = train_raw['Headline']
train['text_b'] = train_raw['articleBody']
```

```python
stances = train_raw['Stance']
stances.replace('agree', 0,True)
stances.replace('disagree', 1, True)
stances.replace('discuss', 2, True)
stances.replace('unrelated', 3, True)
stances
train['labels'] = stances
```

```python
train
```

```python
stances = test_raw['Stance']
stances.replace('agree', 0,True)
stances.replace('disagree', 1, True)
stances.replace('discuss', 2, True)
stances.replace('unrelated', 3, True)
stances
```

```python
test = pd.DataFrame(columns=['text_a', 'text_b'])

test['text_a'] = test_raw['Headline']
test['text_b'] = test_raw['articleBody']
```

```python
test
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
model = ClassificationModel('bert', 'bert-base-uncased', num_labels=4, args={
    'learning_rate':2e-5,
    'num_train_epochs': 5,
    'reprocess_input_data': True,
    'overwrite_output_dir': True,
    'process_count': 10,
    'evaluate_during_training_steps': False,
    'train_batch_size': 4,
    'eval_batch_size': 4,
    'max_seq_length': 512,
    'save_best_model': True,
    'warmup_ratio':0.1,
    'weight_decay':0.01,
    'load_best_model_at_end':True,
    'fp16': False,
    'use_cuda': False,
    'early_stopping' : True
})
```

```python
model.train_model(train)
```

```python
!pwd
```

```python
test_11 = list()
for i in range(len(test)):
  test_11.append([test['text_a'][i], test['text_b'][i]])
```

```python
predictions, raw_outputs = model.predict(test_11)
```

```python
cnt = 0
for i in range(len(predictions)):
  if predictions[i] == stances[i]:
    cnt = cnt + 1
print(cnt/len(predictions))
```

```python
df_answer = pd.DataFrame(columns=['Headline', 'Body ID', 'Stance'])
df_answer['Stance'] = predictions
df_answer.replace(0, 'agree',True)
df_answer.replace(1, 'disagree',True)
df_answer.replace(2, 'discuss',True)
df_answer.replace(3, 'unrelated',True)
df_answer['Headline'] = test_raw['Headline']
df_answer['Body ID'] = test_raw['Body ID']
```

```python
test_update = pd.DataFrame(columns=['Headline', 'Body ID', 'Stance'])
test_update['Stance'] = test_raw['Stance']
test_update.replace(0, 'agree',True)
test_update.replace(1, 'disagree',True)
test_update.replace(2, 'discuss',True)
test_update.replace(3, 'unrelated',True)
test_update['Headline'] = test_raw['Headline']
test_update['Body ID'] = test_raw['Body ID']
```
